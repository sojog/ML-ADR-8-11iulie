# Limitele intre care pot fi plotate elementele

```python
x = np.arange(20)
y = x * 2

plt.plot(x, y)
plt.title("Figura mea")
plt.xlim(0, 6)
plt.ylim(0, 15)
plt.xlabel("Coordonata X")
plt.ylabel("Coordonata Y")

plt.show()
```

```python
a = np.linspace(0, 10, 11)
b = a * 4
a, b
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
axes1.set_title("Prima axa")
axes1.set_xlabel("coord x")
axes1.set_ylabel("coord y")

axes2 = fig.add_axes([0.2, 0.2, 0.5, 0.5])
axes2.plot(a, b)
axes2.set_title("A doua axa")
axes2.set_xlabel("coord x")
axes2.set_ylabel("coord y")

```
