```python
import matplotlib.pyplot as plt
import numpy as np
from  scipy import stats
```

```python
apartamente = [(45, 65),  (55, 70), (60, 82), (70, 90), (90, 130)]


suprafete = [ i[0] for i in apartamente]
preturi =  [ i[1] for i in apartamente]

plt.xlabel("Suprafata (mp)")
plt.ylabel("Pret (mii de euro)")
plt.scatter(suprafete, preturi, color="r")

slope, intercept, *restul = stats.linregress(suprafete, preturi)

plt.plot(suprafete, list(map(lambda x: slope * x + intercept, suprafete)), color="black")

plt.show()
```

```python

a = 0
b = 0
def functie_prezicere(x):
    return a * x + b

plt.plot(suprafete, list(map(functie_prezicere, suprafete)), color="blue")
plt.plot(suprafete, list(map(lambda x: slope * x + intercept, suprafete)), color="black")
plt.scatter(suprafete, preturi, color="r")
plt.show()
```
