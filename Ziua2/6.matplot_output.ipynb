```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.bar(x, y)
plt.show()
```

```python
labels = ["mic", "marisor", "mare", "cel mai mare"]
y = [10, 20, 30, 40]

plt.bar(labels, y)
plt.show()
```

```python
labels = ["mic", "marisor", "mare", "cel mai mare"]
y = [10, 20, 30, 40]

plt.barh(labels, y)
plt.show()
```

```python
labels = ["mic", "marisor", "mare", "cel mai mare"]
date = [10, 20, 30, 40]

plt.pie(date,labels=labels)
plt.show()
```

```python
labels = ["mic", "marisor", "mare", "cel mai mare"]
date = [10, 20, 30, 40]

plt.scatter(date,labels)
plt.show()
```
