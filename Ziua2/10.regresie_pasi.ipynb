```python
import matplotlib.pyplot as plt
import numpy as np
from  scipy import stats
import time
```

```python
apartamente = [(45, 65),  (55, 70), (60, 82), (70, 90), (90, 130)]


suprafete = [ i[0] for i in apartamente]
preturi = preturi_reale =  [ i[1] for i in apartamente]
```

```python
slope, intercept, *restul = stats.linregress(suprafete, preturi)

```

```python

a = 0
b = 0
```

```python
plt.plot(suprafete, list(map(lambda x: a * x + b, suprafete)), color="blue")
plt.scatter(suprafete, preturi, color="r")
plt.show()
```

```python
np_preturi_reale = np.array(preturi_reale) 
np_suprafete = np.array(suprafete)


def calculeaza(m=len(suprafete), learning_rate = 0.0001):
    global a, b
    # print(f"La inceput: a={a}, b={b}") 

    preturi_prezise = np.array(list(map(lambda x: a*x+b, suprafete)))

    cost = 1/(2*m) * np.sum((np_preturi_reale - preturi_prezise) ** 2) 
    
    gradient_a = 1/m * np.sum((preturi_prezise - np_preturi_reale ) * np_suprafete)

    gradient_b = 1/m * np.sum((preturi_prezise - np_preturi_reale ) ) * 1

    a = a - gradient_a * learning_rate

    b = b - gradient_b * learning_rate 

    print("gradient_A", gradient_a)
    print("gradient_B", gradient_b)

    return a, b, cost




```

```python
print(calculeaza())
plt.plot(suprafete, list(map(lambda x: a * x + b, suprafete)), color="blue")
plt.scatter(suprafete, preturi, color="r")
plt.plot(suprafete, list(map(lambda x: slope * x + intercept, suprafete)), color="black")
plt.show()
```

```python
for i in range(10001):
    calculeaza()


plt.plot(suprafete, list(map(lambda x: a * x + b, suprafete)), color="blue")
plt.scatter(suprafete, preturi, color="r")
plt.plot(suprafete, list(map(lambda x: slope * x + intercept, suprafete)), color="black")
plt.show()
time.sleep(5)
        
```
