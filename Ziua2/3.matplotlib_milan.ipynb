```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os
```

```python
image_files = os.listdir("Milan2007")
```

```python
type(image_files)
```

```python
image_files.sort()
```

```python
image_files = image_files[:11]
```

```python
figure, axis = plt.subplots(nrows=1, ncols=11)

for index, image_name in enumerate(image_files):
    image = mpimg.imread(f"Milan2007/{image_name}")
    axis[index].imshow(image)
```

```python
figure, axis = plt.subplots(nrows=1, ncols=11)

for index, image_name in enumerate(image_files):
    image = mpimg.imread(f"Milan2007/{image_name}")
    axis[index].imshow(image)
    axis[index].axis("off")
```

```python
figure, axis = plt.subplots(nrows=5, ncols=7)

for row in range(5):
    for column in range(7):
        axis[row][column].axis("off")

player_indexes = [(4, 3),
                  (3, 0), (3, 2), (3, 4), (3, 6),
                  (2, 1), (2, 3), (2, 5),
                  (1, 2), (1, 4),
                  (0, 3)]


for index, image_name in enumerate(image_files):
    image = mpimg.imread(f"Milan2007/{image_name}")
    player_index = player_indexes[index]
    axis[player_index[0], player_index[1]].imshow(image)
    axis[player_index[0], player_index[1]].axis("off")
```
