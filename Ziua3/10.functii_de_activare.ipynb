```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 3)
x
```

```python
x = np.linspace(-10, 10, 10)
x
```

```python
x = np.linspace(-5, 5, 100)
y_sigmoid = sigmoid(x)
```

```python
plt.xlabel("x")
plt.ylabel("Sigmoid y")
plt.title("Functia Sigmoidă")

plt.plot(x, y_sigmoid, label="Sigmoid")

# plt.axhline(y=0.5)
# plt.axvline(x=0)
plt.show()
```

```python

def relu(x):
    return 0 if x < 0 else x 
```

```python
plt.xlabel("x")
plt.ylabel("RELU y")
plt.title("Functia RELU")

plt.plot(x, list(map(relu, x)), label="RELU")

# plt.axhline(y=0.5)
# plt.axvline(x=0)
plt.show()
```
