```python
image = cv2.imread("plane.jpg", cv2.IMREAD_GRAYSCALE)
blurred_image = cv2.blur(image, (50, 50))

plt.imshow(blurred_image, cmap="gray")
plt.axis("off")
plt.show()
```

# Creare kernel

```python
import numpy as np
```

```python
kernel = np.ones((3, 3))
kernel
```

```python
image = cv2.imread("plane.jpg", cv2.IMREAD_GRAYSCALE)
kernel_applied_image = cv2.filter2D(image,-1, kernel)
plt.imshow(kernel_applied_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
SIZE = 5

kernel = np.ones((SIZE, SIZE)) / 15.0
print(kernel)

image = cv2.imread("plane.jpg", cv2.IMREAD_GRAYSCALE)
kernel_applied_image = cv2.filter2D(image,-1, kernel)
plt.imshow(kernel_applied_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
setosa_blur_kernel = [0.0625,
0.125,
0.0625,
0.125,
0.25,
0.125,
0.0625,
0.125,
0.0625,]

setosa_blur_kernel = np.array(setosa_blur_kernel).reshape(3, 3)
setosa_blur_kernel
```

```python
image = cv2.imread("plane.jpg", cv2.IMREAD_GRAYSCALE)
kernel_applied_image = cv2.filter2D(image,-1, setosa_blur_kernel)
plt.imshow(kernel_applied_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
image = cv2.imread("small_img2.jpg", cv2.IMREAD_GRAYSCALE)
kernel_applied_image = cv2.filter2D(image,-1, setosa_blur_kernel)
plt.imshow(kernel_applied_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
top_kernel = np.array([[1, 2,1],
       [0 , 0  , 0 ],
       [-1, -2 , -1]])

image = cv2.imread("small_img2.jpg", cv2.IMREAD_GRAYSCALE)
kernel_applied_image = cv2.filter2D(image,-1, top_kernel)
plt.imshow(kernel_applied_image, cmap="gray")
plt.axis("off")
plt.show()
```

```python
left_kernel = np.array([[1, 2,1],
       [0 , 0  , 0 ],
       [-1, -2 , -1]]).T
print(left_kernel)
image = cv2.imread("small_img2.jpg", cv2.IMREAD_GRAYSCALE)
kernel_applied_image = cv2.filter2D(image,-1, left_kernel)
plt.imshow(kernel_applied_image, cmap="gray")
plt.axis("off")
plt.show()
```
